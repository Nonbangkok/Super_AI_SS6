รายละเอียดชุดข้อมูล (Dataset Description)
ชุดข้อมูลนี้แสดง สถิติหน่วยจำหน่ายไฟฟ้า (Electricity Consumption) ของพื้นที่ให้บริการของ การไฟฟ้านครหลวง (MEA) โดยจำแนกตาม เขตการจำหน่ายไฟฟ้า และช่วงเวลา รายเดือน

# Cleaning Dataset


```python
# strip leading/trailing spaces from all column names
df.columns = df.columns.str.strip()
df.columns
```

```python
# strip leading/trailing spaces from df['เดือน']
df['เดือน'] = df['เดือน'].str.strip()
df['เดือน']
```

```python
# fill missing values in the 'ปี' (year) column by propagating the last valid observation forward.
df['ปี'] = df['ปี'].fillna(method='ffill')
df['ปี']
```

```python
# convert datatype of df['ปี'] from float64 to int64
df['ปี'] = df['ปี'].astype(int)
df.info()
```

```python
# mapping dict
month_map = {
    'ม.ค.': 1, 'ก.พ.': 2, 'มี.ค.': 3, 'เม.ย.': 4,
    'พ.ค.': 5, 'มิ.ย.': 6, 'ก.ค.': 7, 'ส.ค.': 8,
    'ก.ย.': 9, 'ต.ค.': 10, 'พ.ย.': 11, 'ธ.ค.': 12
}

# mapping all month to correspond number
df['month_num'] = df['เดือน'].map(month_map).astype(int)
df['month_num']
```

```python
# create df['data_date']
df['year_ad'] = df['ปี'] - 543 # Convert Buddhist year to Gregorian year
df['data_date'] = pd.to_datetime(df['year_ad'].astype(str) + '-' + df['month_num'].astype(str).str.zfill(2))
df['data_date']
```

```python
# drop unused column
df = df.drop(columns='(หักไฟทำการ)')
```

```python
# Unpivot the DataFrame from wide to long format,
# consolidating district columns into a single 'district' variable
# and their values into 'electricity_unit'.
id_vars = ['ปี', 'เดือน', 'month_num', 'year_ad', 'data_date']
value_vars = [col for col in df.columns if col not in id_vars]

df_long = pd.melt(df, id_vars=id_vars, value_vars=value_vars, var_name='district', value_name='electricity_unit')

print("First 5 rows of the new long-format DataFrame:")
print(df_long.head())
print("\nInformation about the new long-format DataFrame:")
print(df_long.info())
```

## Electricity Usage Across Different Districts Over Time

```python
import altair as alt

# Create Electricity Usage Across Different Districts Over Time Chart
chart = alt.Chart(df_long).mark_line().encode(
    x=alt.X('data_date', title='Date'),
    y=alt.Y('electricity_unit', title='Electricity Unit (kWh)'),
    color='district',
    tooltip=['data_date', 'district', 'electricity_unit']
).properties(
    title='Electricity Usage Across Different Districts Over Time'
).interactive()

chart.display()
```

กราฟเส้นแสดงปริมาณการใช้ไฟฟ้าจำแนกตามเขตของกลุ่มตัวอย่างในพื้นที่การไฟฟ้านครหลวงระหว่างปี 2019 ถึง 2025 สะท้อนภาพรวมพฤติกรรมการใช้พลังงานที่เคลื่อนไหวไปในทิศทางเดียวกันอย่างมีนัยสำคัญ โดยเห็นแนวโน้มการเติบโตแบบค่อยเป็นค่อยไปควบคู่ไปกับรูปแบบฤดูกาล (Seasonality) ที่ชัดเจนมาก โดยเฉพาะช่วงกลางของแต่ละปีซึ่งมักเป็นฤดูร้อนที่จะเห็นจุดยอดของการใช้ไฟฟ้าพุ่งสูงขึ้นเนื่องจากการทำงานของระบบทำความเย็น ขณะที่ปี 2020-2021 ปรากฏการชะลอตัวและผันผวนเล็กน้อยซึ่งอาจเชื่อมโยงกับภาวะเศรษฐกิจในช่วงวิกฤตสุขภาพที่ผ่านมา อย่างไรก็ตาม ความแตกต่างของระดับเส้นกราฟบ่งบอกถึงลักษณะทางกายภาพที่ต่างกันของแต่ละเขต เช่น บางพลีและคลองเตยที่รักษาฐานการใช้ไฟฟ้าในระดับสูงจากกิจกรรมทางอุตสาหกรรมและพาณิชย์ขนาดใหญ่ ตรงข้ามกับเขตวัดเลียบที่ใช้พลังงานน้อยกว่าอย่างเห็นได้ชัด ภาพรวมในปี 2025 แสดงให้เห็นถึงการขยายตัวของเมืองที่ยังคงดำเนินต่อเนื่อง สะท้อนผ่านความต้องการพลังงานที่เพิ่มขึ้นในเกือบทุกเขตพื้นที่อย่างมั่นคงและเป็นระบบตามการขยายตัวของโครงสร้างพื้นฐานเขตเมือง

## Average Monthly Electricity Usage Pattern

```python
# Calculate the average electricity usage per month
monthly_avg_usage = df_long.groupby('month_num')['electricity_unit'].mean().reset_index()

# Map month numbers back to names for better readability on the plot
month_names_map = {v: k for k, v in month_map.items()}
monthly_avg_usage['month'] = monthly_avg_usage['month_num'].map(month_names_map)

# Define the order of months for the x-axis
month_order = list(month_map.keys())

# Create the line chart for monthly electricity usage patterns
chart_monthly_pattern = alt.Chart(monthly_avg_usage).mark_line().encode(
    x=alt.X('month', sort=month_order, title='Month'),
    y=alt.Y('electricity_unit', title='Average Electricity Unit (kWh)'),
    tooltip=['month', 'electricity_unit']
).properties(
    title='Average Monthly Electricity Usage Pattern'
).interactive()

chart_monthly_pattern.display()
```

กราฟแสดงรูปแบบฤดูกาลที่ชัดเจนโดยค่าเฉลี่ยการใช้ไฟฟ้าพุ่งสูงที่สุดในเดือนพฤษภาคมซึ่งสัมพันธ์โดยตรงกับอุณหภูมิที่ร้อนจัดในประเทศไทย ส่งผลให้ความต้องการพลังงานเพื่อระบายความร้อนเพิ่มขึ้น ก่อนจะค่อยๆ ลดระดับลงในช่วงปลายปีตามสภาพอากาศที่เย็นลง สะท้อนให้เห็นว่าปัจจัยทางภูมิอากาศมีอิทธิพลสำคัญต่อพฤติกรรมการใช้ไฟฟ้าในเขตกรุงเทพและปริมณฑลมากกว่าปัจจัยด้านอื่นอย่างมีนัยสำคัญ

## Average Electricity Usage by District

```python
import pandas as pd

# Calculate the average electricity usage per district
district_avg_usage = df_long.groupby('district')['electricity_unit'].mean().reset_index()

# Sort districts by average electricity usage in descending order
district_avg_usage_ranked = district_avg_usage.sort_values(by='electricity_unit', ascending=False).reset_index()

print("Average electricity usage by district (ranked):")
print(district_avg_usage_ranked)
```

```python
# Create a bar chart for average electricity usage by district
chart_district_rank = alt.Chart(district_avg_usage_ranked).mark_bar().encode(
    x=alt.X('electricity_unit', title='Average Electricity Unit (kWh)'),
    y=alt.Y('district', sort='-x', title='District'),
    tooltip=['district', 'electricity_unit']
).properties(
    title='Average Electricity Usage by District'
).interactive()

chart_district_rank.display()
```

แผนภูมิแท่งแสดงระดับการใช้ไฟฟ้าเฉลี่ยรายเขต พบว่าบางพลีและสมุทรปราการครองอันดับสูงสุดอย่างโดดเด่นสะท้อนถึงอิทธิพลของภาคอุตสาหกรรมและพาณิชย์ขนาดใหญ่ในพื้นที่ ขณะที่เขตวัดเลียบมีการใช้น้อยที่สุดอย่างเห็นได้ชัด ข้อมูลนี้บ่งชี้ถึงความเหลื่อมล้ำของการใช้พลังงานตามลักษณะกิจกรรมทางเศรษฐกิจและความหนาแน่นของแต่ละเขตพื้นที่การปกครองอย่างชัดเจน

## District Growth Rates

```python
# Calculate average electricity usage per district in 2019
avg_usage_2019 = df_long[df_long['year_ad'] == 2019].groupby('district')['electricity_unit'].mean().reset_index()
avg_usage_2019 = avg_usage_2019.rename(columns={'electricity_unit': 'avg_unit_2019'})
print("Average electricity usage per district in 2019:")
print(avg_usage_2019.head())
```

```python
# Calculate average electricity usage per district in 2025
avg_usage_2025 = df_long[df_long['year_ad'] == 2025].groupby('district')['electricity_unit'].mean().reset_index()
avg_usage_2025 = avg_usage_2025.rename(columns={'electricity_unit': 'avg_unit_2025'})
print("Average electricity usage per district in 2025:")
print(avg_usage_2025.head())
```

```python
# Merge the df between 2019 and 2025
merged_usage = pd.merge(avg_usage_2019, avg_usage_2025, on='district')
print("Merged DataFrame showing average usage for 2019 and 2025:")
print(merged_usage.head())
```

```python
# Calculate percentage growth rates
merged_usage['growth_rate_percent'] = ((merged_usage['avg_unit_2025'] - merged_usage['avg_unit_2019']) / merged_usage['avg_unit_2019']) * 100
print("District growth rates (2019-2025):")
print(merged_usage.head())
```

## STL Decomposition

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import FuncFormatter
import seaborn as sns
from statsmodels.tsa.seasonal import STL

# --- 1. Data Preparation ---
# Ensure 'data_date' is datetime and aggregate units by month
df_long['data_date'] = pd.to_datetime(df_long['data_date'])
df_total = df_long.groupby('data_date')['electricity_unit'].sum().reset_index()
df_total.set_index('data_date', inplace=True)
df_total = df_total.asfreq('MS')

# --- 2. STL Decomposition ---
# Use robust=True to handle outliers and seasonal=13 for monthly cycles
stl = STL(df_total['electricity_unit'], seasonal=13, robust=True)
res = stl.fit()

# --- 3. Anomaly Detection ---
residual = res.resid
threshold = 2 * residual.std()
df_total['residual'] = residual
anomalies = df_total[np.abs(df_total['residual']) > threshold]

# --- 4. Custom Formatter for Thai Buddhist Era (พ.ศ.) ---
# This function converts Year (AD) to Year (BE) by adding 543
def to_be_year(x, pos):
    dt = mdates.num2date(x)
    return f"{dt.year + 543}"

be_formatter = FuncFormatter(to_be_year)

# --- 5. Visualization ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(14, 16), sharex=True)

# Subplot 1: Observed Data & Anomalies
ax1.plot(df_total.index, df_total['electricity_unit'], label='Total Observed (MEA)', color='#2c3e50', linewidth=2)
ax1.scatter(anomalies.index, anomalies['electricity_unit'], color='#e74c3c', s=60, label='Anomalies', zorder=5)
ax1.set_title('1. Electricity Consumption & Anomaly Detection (2562-2568 BE)', fontsize=14, fontweight='bold')
ax1.set_ylabel('Million Units (GWh)')
ax1.legend(loc='upper left')

# Subplot 2: Trend Component
ax2.plot(res.trend, label='Trend', color='#27ae60', linewidth=2)
ax2.set_title('2. Long-term Trend (Economic & Urban Growth)', fontsize=12)
ax2.set_ylabel('Trend (GWh)')

# Subplot 3: Seasonal Component
ax3.plot(res.seasonal, label='Seasonality', color='#f39c12', linewidth=1.5)
ax3.set_title('3. Seasonal Patterns (Temperature & Climate Effect)', fontsize=12)
ax3.set_ylabel('Seasonal (GWh)')

# Subplot 4: Residuals
ax4.bar(df_total.index, residual, label='Residuals', color='#95a5a6', alpha=0.7, width=20)
ax4.axhline(threshold, color='#c0392b', linestyle='--', linewidth=1, label='Threshold (2σ)')
ax4.axhline(-threshold, color='#c0392b', linestyle='--', linewidth=1)
ax4.set_title('4. Residuals (Tracking: COVID, Heatwave, Policy)', fontsize=12)
ax4.set_ylabel('Residual Value')
ax4.legend(loc='upper left')

# --- Applying Thai Buddhist Era Year Labels ---
for ax in [ax1, ax2, ax3, ax4]:
    ax.xaxis.set_major_locator(mdates.YearLocator())
    ax.xaxis.set_major_formatter(be_formatter)

plt.xlabel('Year (B.E.)', fontsize=12)
plt.tight_layout()
plt.show()

# --- Anomaly Report ---
print("--- Anomaly Detection Report (B.E. Years) ---")
if not anomalies.empty:
    # Display index with +543 year adjustment for clarity in text
    report_df = anomalies[['electricity_unit', 'residual']].copy()
    print(report_df)
else:
    print("No significant anomalies detected.")
```

1. Electricity Consumption & Anomaly Detection
กราฟแรกแสดงภาพรวมปริมาณการใช้ไฟฟ้าจริงที่เกิดขึ้น โดยจุดสีแดงคือ "จุดผิดปกติ" (Anomalies) ที่เบี่ยงเบนออกจากรูปแบบปกติเกิน 2 ส่วนเบี่ยงเบนมาตรฐาน พบความผิดปกติเด่นชัดในช่วงต้นปี 2563 ซึ่งสอดคล้องกับการเริ่มต้นของสถานการณ์ COVID-19 ที่เปลี่ยนพฤติกรรมการใช้พลังงานอย่างฉับพลัน และอีกระลอกในช่วงต้นปี 2567 ซึ่งอาจเกิดจากสภาวะเอลนีโญที่ทำให้อากาศร้อนจัดรุนแรงกว่าปกติ จุดเหล่านี้สะท้อนให้เห็นว่าระบบไฟฟ้าในเขตเมืองมีความอ่อนไหวสูงต่อปัจจัยภายนอกที่ไม่คาดคิด ทั้งด้านสาธารณสุขและวิกฤตสภาพภูมิอากาศ

2. Long-term Trend
กราฟเทรนด์ช่วยขจัดปัจจัยฤดูกาลออกไปเพื่อดูทิศทางเศรษฐกิจเมือง พบว่าในช่วงปี 2563–2564 แนวโน้มการใช้ไฟฟ้าดิ่งลงอย่างต่อเนื่อง สะท้อนภาวะเศรษฐกิจชะลอตัวจากการล็อกดาวน์และการหยุดชะงักของภาคบริการ อย่างไรก็ตาม ตั้งแต่กลางปี 2564 เป็นต้นมา เทรนด์กลับมาฟื้นตัวเป็นขาขึ้นอย่างแข็งแกร่งและต่อเนื่องจนถึงปี 2567 สอดคล้องกับการกลับมาของกิจกรรมทางเศรษฐกิจและการท่องเที่ยว แม้ในช่วงปลายปี 2567 จะเริ่มเห็นการชะลอตัวลงเล็กน้อย แต่ภาพรวมยังคงอยู่ในระดับที่สูงกว่าช่วงก่อนเกิดโรคระบาด บ่งบอกถึงการขยายตัวของโครงสร้างพื้นฐานและจำนวนผู้อยู่อาศัยในเขตเมืองที่เพิ่มขึ้น

3. Seasonal Patterns
รูปแบบฤดูกาลแสดงความสม่ำเสมอของวงจรการใช้ไฟฟ้าในแต่ละปีอย่างชัดเจน โดยทุกปีจะเกิด "Peak Period" ในช่วงเดือนเมษายนถึงพฤษภาคม ซึ่งเป็นช่วงที่อุณหภูมิในประเทศไทยสูงที่สุด ส่งผลให้ความต้องการใช้ไฟฟ้าเพื่อการปรับอากาศพุ่งสูงขึ้นเป็นประวัติการณ์ ในขณะที่ช่วงปลายปี (พฤศจิกายน-มกราคม) การใช้ไฟฟ้าจะลดต่ำลงตามสภาพอากาศที่เย็นลง รูปแบบที่ซ้ำเดิมทุกปีนี้ช่วยให้การไฟฟ้านครหลวงสามารถวางแผนการสำรองพลังงานได้อย่างแม่นยำ และชี้ให้เห็นว่าพฤติกรรมการใช้ไฟฟ้าของคนเมืองผูกติดกับดัชนีอุณหภูมิอย่างมีนัยสำคัญที่สุด

4. Residuals
กราฟส่วนที่เหลือ (Residuals) คือค่าที่เหลือจากการตัดเทรนด์และฤดูกาลออกไป เพื่อใช้ระบุเหตุการณ์เฉพาะกิจที่ส่งผลกระทบชั่วคราว เส้นประสีแดงแสดงเกณฑ์การตรวจจับความผิดปกติ ซึ่งจะเห็นว่ามีแท่งกราฟหลายแท่งพุ่งสูงเกินเกณฑ์อย่างมีนัยสำคัญ โดยเฉพาะในช่วงมกราคมของปี 2563 และ 2567 ข้อมูลส่วนนี้มีค่าเป็นบวกสูงหมายความว่ามีการใช้ไฟฟ้าเกินกว่าที่โมเดลคาดการณ์ไว้มาก ซึ่งเชื่อมโยงกับคลื่นความร้อนที่มาเร็วกว่ากำหนด การแพร่ละบาดของ COVID-19 หรือมาตรการกระตุ้นเศรษฐกิจของรัฐบาลที่ทำให้เกิดกิจกรรมในเขตเมืองหนาแน่นขึ้นในช่วงเวลาดังกล่าว

## Answering The Question

⚡ ข้อ 1: แนวโน้มและการตรวจจับจุดผิดปกติ (Trend & Anomaly)
การใช้ไฟฟ้ารวมเปลี่ยนแปลงอย่างไรใน 7 ปี?

แนวโน้ม (Trend): พบว่าการใช้ไฟฟ้าของกรุงเทพฯ และปริมณฑลผ่านจุดต่ำสุดในช่วงปี 2564 ซึ่งเป็นช่วงวิกฤต COVID-19 ที่เข้มงวดที่สุด หลังจากนั้นมีทิศทาง ฟื้นตัวเป็นขาขึ้นอย่างต่อเนื่อง (V-Shape Recovery) จนถึงปี 2568 สะท้อนให้เห็นถึงการฟื้นตัวของกิจกรรมทางเศรษฐกิจและการขยายตัวของเมืองที่หนาแน่นขึ้น

รูปแบบฤดูกาล (Seasonality): กราฟแสดง "Peak" ที่ชัดเจนในเดือน เมษายน-พฤษภาคม ของทุกปี โดยมีส่วนต่างจากค่าเฉลี่ยปกติสูงถึง 400-500 GWh บ่งบอกว่าสภาพภูมิอากาศเช่น อุณหภูมิ มีผลต่อโหลดไฟฟ้ามากกว่ากิจกรรมทางเศรษฐกิจทั่วไป

จุดผิดปกติ (Anomaly): ตรวจพบจุดผิดปกติในต้นปี 2563 (COVID Lockdown) ที่ค่า Residual พุ่งสูงและต่ำลง และต้นปี 2567 (Extreme Heatwave) ที่การใช้ไฟฟ้าพุ่งสูงเกินคาดการณ์

ทิศทางปี 2568: แม้แนวโน้มจะยังเป็นบวก แต่เริ่มเห็นการ "Plateau" หรือทรงตัว ในระดับสูง ซึ่งสะท้อนว่าภาวะเศรษฐกิจเมืองเข้าสู่ระยะอิ่มตัวหรือมีการใช้เทคโนโลยีประหยัดพลังงานมาช่วยควบคุมโหลดไฟฟ้ามากขึ้น

🏙 ข้อ 2: การจัดอันดับและการเติบโตรายเขต (Ranking & Growth)
เขตไหนคือผู้นำ และใครคือดาวรุ่ง?
เราสามารถแบ่งเขตออกเป็น 3 กลุ่มหลักตามพฤติกรรมการเติบโต:

High Volume, Steady Growth (ยักษ์ใหญ่คงที่): ได้แก่ บางพลี และ สมุทรปราการ เขตเหล่านี้ใช้ไฟมากที่สุด (Peak > 450 GWh) เพราะเป็นเขตอุตสาหกรรมและคลังสินค้าหลักของประเทศ

Low Volume, High Growth (ดาวรุ่งพุ่งแรง): ได้แก่ บางบัวทอง และ บางใหญ่ ที่มีอัตราการเติบโตพุ่งสูงกว่า 25% สะท้อนถึงการขยายตัวของโครงการหมู่บ้านจัดสรรตามแนวรถไฟฟ้าสายสีม่วง (Residential Expansion)

Negative Growth (เขตที่อิ่มตัว): ได้แก่ วัดเลียบ และ ราชบูรณะ ที่การใช้ไฟฟ้าลดลงจาก 7 ปีที่แล้ว บ่งบอกถึงการเคลื่อนย้ายของประชากรหรือการเปลี่ยนผ่านจากย่านการค้าเก่าไปสู่พื้นที่เศรษฐกิจใหม่

🌦 ข้อ 3: รูปแบบฤดูกาลรายเขต (Seasonal Pattern by District)
ความผันผวนต่างกันอย่างไร?
จากการวิเคราะห์ค่าสัมประสิทธิ์ความผันผวน (CV) เราพบสมมติฐานที่ชัดเจนดังนี้:

เขตที่ผันผวนสูง (High Volatility): เช่น บางใหญ่ และ นวลจันทร์ มีกราฟพุ่งโด่งในฤดูร้อนและดิ่งต่ำในฤดูหนาว ชี้ชัดว่าเป็น "เขตที่พักอาศัย (Residential)" ที่โหลดไฟฟ้าขึ้นอยู่กับเครื่องปรับอากาศเป็นหลัก

เขตที่คงที่ (Stable Load): เช่น คลองเตย และ บางพลี มีเส้นกราฟที่ค่อนข้างราบเรียบตลอดทั้งปี (CV ต่ำที่สุด) เนื่องจากเป็น "เขตพาณิชย์/อุตสาหกรรม" ที่มีการเดินเครื่องจักร คลังสินค้าแช่เย็น หรือสำนักงานที่เปิดทำงานสม่ำเสมอไม่ว่าฤดูใด

จุดสังเกตพิเศษ: ในเดือนเมษายนเกือบทุกเขตจะมีการย่อตัวเล็กน้อย (Minor Drop) ซึ่งอธิบายได้จาก "เทศกาลสงกรานต์" ที่โรงงานและสำนักงานหยุดทำการยาว ทำให้โหลดไฟฟ้าจากภาคธุรกิจลดลงสวนทางกับความร้อน

บทสรุปสำหรับ Hackathon: จากการวิเคราะห์เชิงลึก จะพบประเด็นในหลาย ๆ แง่ เช่น "การขยายตัวของเมืองสู่ทิศตะวันตก (บางบัวทอง/บางใหญ่)" และ "ความเปราะบางของโหลดไฟฟ้าต่อสภาพภูมิอากาศในเขตที่อยู่อาศัย" ซึ่งจะเป็น Insight ที่ทรงพลังในการทำความเข้าใจโครงสร้างการใช้พลังงานและวางเชิงนโยบาย
